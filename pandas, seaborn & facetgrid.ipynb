## Lectura de los datos

```python
data = pd.read_csv('Seattle2014.csv')
data.head(5)
```

```python
print(f"El fichero Seattle2014.csv tiene {data.shape[0]} filas y {data.shape[1]} columnas.")
data.dtypes
```

Con la función `describe()` podemos resumir las variables numéricas del `DataFrame`. Los estadísticos de resumen son:

  * `count`: Número de casos. Útil para ver si hay casos perdidos
  * `mean`: Valor promedio de las observaciones
  * `std`: Desviación estándar de los datos
  * `min`: Valor mínimo observado
  * `25%`: Percentil 25 de las observaciones
  * `50%`: Percentil 50 de las observaciones. Coincide con el valor de la mediana
  * `75%`: Percentil 75 de las observaciones
  * `max`: Valor máximo observado
  
El percentil es una medida de posición usada en estadística que indica, una vez ordenados los datos de menor a mayor, el valor de la variable por debajo del cual se encuentra un porcentaje dado de observaciones en un grupo de observaciones. Por ejemplo, el percentil 20$ es el valor bajo el cual se encuentran el 20 por ciento de las observaciones.

```python
data.describe()
```

### Eliminar columnas (_dropping columns_)

A menudo, podemos encontrarnos con conjuntos de datos con muchas variables y no todas útiles o interesantes. Por ejemplo, en el caso de los datos que nos ocupa, vemos que la variable **STATION** y **STATION_NAME** es constante para los 365 registros.

En este caso, vamos a eliminar las variables de **AWND** hasta **WT03**.

`pandas` proporciona una forma práctica de eliminar columnas o filas no deseadas de un `DataFrame` con la función `drop()`. En el código, creamos la variable `a_eliminar`que tiene la lista de variables que queremos quitar del `DataFrame`. Con `data.drop()` las eliminamos con los siguientes argumentos:

  * `columns`: Lista de variables a eliminar
  * `inplace`: Argumento muy útil. Si es `True` el `DataFrame` se actualiza sin esas columnas. Si es `False` presenta los datos pero el `DataFrame`no cambia
  * `axis`: Indica que que se elimina por columnas (`axis = 1`) 

En este caso, la información los parámetros son redundantes. Al poner `columns`la función ya sabe que tiene que eliminar columnas, así que el parámetro `axis = 1` sobraría.

```python
a_eliminar = ['AWND', 'WDF2', 'WDF5', 'WSF2', 'WSF5', 'WT01', 'WT05', 'WT02', 'WT03']
data.drop(columns = a_eliminar, inplace = False, axis = 1).head(5)
```

### Cambiar el índice de un `DataFrame`

Un Índices en `pandas` amplía la funcionalidad de las matrices `NumPy` para permitir un corte y etiquetado de las observaciones mucho más versátil. En muchos casos, es útil usar un campo de identificación con valor único como su índice.

Por ejemplo, en el conjunto de datos que estamos utilizando, se puede esperar que el identificador sea la fecha de registro, ya que es una única fecha por observación (365 registros, uno por cada día del año).

Con la función `is_unique` validamos que efectivamente la variable tenga un valor distinto en cada registro. Utilizamos la función `set_index()` para añadir como índice del `DataFrame`la variable pasada como argumento. Esta función también permite el argumento `inplace` por lo que no tendríamos que hacer una asignación.

```python
data2 = data.copy()
data2.drop(columns = a_eliminar, inplace = True)
data2.head(5)
```

```python
data2['DATE'].is_unique
```

```python
data2 = data2.set_index('DATE')
data2.head(5)
```

Ahora, es mucho más fácil acceder a la información del `DataFrame` a partir del índice. Por ejemplo, si queremos conocer los registros entre los días 8 de junio al 15 de junio de 2014 en Seattle utilizamos la función `.loc` para acceder a los registros con índice [20140608, 20140609, 20140610, 20140611, 20140612, 20140613, 20140614, 20140615].

```python
data2.loc[range(20140608,20140615)]
```

### Poniendo orden (_tydying_)

Hasta el momento, sólo hemos eliminado columnas innecesarias y hemos cambiado el índice del `DataFrame` a algo más útil. En este punto, vamos a limpiar columnas específicas y obtener un formato uniforme para obtener una mejor comprensión del conjunto de datos. En particular, limpiaremos la fecha de registro (`datetime`).

Tras la inspección, casi todos los datos son valores numéricos (incluida la fecha de registro **DATE**) y dos cadenas de caracteres (**STATION** y **STATION_NAME**).

```python
data.get_dtype_counts()
```

#### Trabajo con fechas

La variable **DATE** está en formato yyymmdd pero en formato numérico. Con el comando `dtypes` podemos comprobarlo. Al ser númerico tengo dos opciones:

1. Trabajar como número y extraer la secuencia del valor del mes o
2. Pasar a `string` y trabajar con el paquete `datetime`

En este ejemplo vamos a trabajar con el paquete `datetime`.

```python
data['DATE'] = data['DATE'].astype(str)
data.dtypes
```

Utilizamos la función `to_datetime` y convertimos a fecha la cadena de texto almacenada en **DATE** con formato `%Y%m%d`. En la siguiente tabla podemos observar que la variable **DATE** ya se representa de distinta manera.

```python
data['DATE'] =  pd.to_datetime(data['DATE'], format='%Y%m%d')
data.head(5)
```

Ahora podemos usar el paquete `datetime` para sacar la información de la fecha. Con `dt.month` calculamos el mes de la fecha en cuestión, un valor de 1 a 12.

```python
data['MONTH'] = data['DATE'].dt.month
data.head(5)
```

Podíamos haberlo hecho todo en una sóla línea, utilizando el siguiente comando:

`data['MONTH'] = pd.to_datetime(data['DATE'], format = '%Y%m%d').dt.month`

### Análisis descriptivos

La variable `data`contiene la información de la temperatura mínima `TMIN` y temperatura máxima `TMAX` de cada día del año. Calculamos el valor promedio de ambas variables por mes.

```python
data.groupby('MONTH')['TMIN', 'TMAX'].mean()
```

Un análisis mucho más exhaustivo podíamos haberlo generado con la función `describe()`.

## Gráficos

### Con `pandas.boxplot()`

Mi intención es generar un gráfico _boxplot_ en el que se muestre la distribución de `TMIN` y `TMAX` por mes. Un primer resultado lo podemos obtener mediante `pandas.boxplot`.

```python
data.boxplot(column = ['TMIN','TMAX'], by = 'MONTH');
```

Es una forma rápida de visualizar la distribución de una variable en función de otra (en este caso `MONTH`). Si quisiera agrupar la distribución de la temperatura mínima y máxima registrada usando el mismo comando tendría que modificar los datos. Podría utilizar las función `melt()` y ` gather()`, pero por comodidad voy a usar la función `pandas.concat()`.

```python
aux1 = pd.concat([data[['TMIN']].melt(), data[['MONTH']]], axis = 1)
aux2 = pd.concat([data[['TMAX']].melt(), data[['MONTH']]], axis = 1)
aux3 = pd.concat([aux1, aux2])
aux3.head(5)
```

He creado un `DataFrame` en el que la columna `variable` contiene la información de si es la temperatura mínima (`TMIN`) o máxima (`TMAX`) la registrada. La columna `value` contiene el dato y la variable `MONTH` el mes en el que se registró.

```python
aux3.boxplot(column = ['value'], by = ['MONTH', 'variable'], figsize = (10, 7));
```

La función `pandas.boxplot()` es muy útil ya que representa rápidamente el comportamiento de las variables estudiadas, pero no he visto que sea muy cómodo mejorar el gráfico de salida. He probado otras opciones, con `seaborn`.

### Con `seaborn`

```python
sns.factorplot(x = 'MONTH', y = 'value',\
        hue = 'variable', data = aux3, \
        kind = 'box', palette = 'Set1', \
        linewidth = 1.0, size = 6, aspect = 2);
```

#### FacetGrid

El paquete `seaborn` tiene una función, `FacetGrid` que permite hacer uan rejilla de gráficos en la que cada fila y columna implique un factor a analizar. En el siguiente ejemplo voy a realizar un histograma con la distribución de la temperatura por mes analizado.

En los siguientes links he encontrado bastante ayuda:

  * https://seaborn.pydata.org/generated/seaborn.FacetGrid.html
  * https://stackoverflow.com/questions/31632372/customizing-annotation-with-seaborns-facetgrid
  * http://seaborn.pydata.org/tutorial/axis_grids.html#mapping-custom-functions-onto-the-grid
  * https://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-styles


```python
g = sns.FacetGrid(aux3, row = 'variable', col = 'MONTH', margin_titles = True)
bins = np.linspace(aux3['value'].min(), aux3['value'].max(), 25)
g.map(plt.hist, 'value', color = 'steelblue', bins = bins, lw = 0)
g.set_titles(col_template = 'Mes: {col_name}', row_template = 'Temperatura: {row_name}');
```

Vemos que con los argumentos `col_template` y `row_template` podemos cambiar los títulos de cada subgráfico en función del mes y temperatura registrada.

```python
aux3['MONTH'] = aux3['MONTH'].map({1: 'Enero', 2: 'Febrero', 3: 'Marzo', \
                                   4: 'Abril', 5: 'Mayo', 6: 'Junio', \
                                   7: 'Julio', 8: 'Agosto', 9: 'Septiembre', \
                                   10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'})
```

```python
g = sns.FacetGrid(aux3, row = 'variable', col = 'MONTH', margin_titles = True)
bins = np.linspace(aux3['value'].min(), aux3['value'].max(), 25)
g.map(plt.hist, 'value', color = 'steelblue', bins = bins, lw = 0)
g.set_titles(col_template = 'Mes: {col_name}', row_template = 'Temperatura: {row_name}');
```

Analizamos ahora sólo los meses, independientemente de la temperatura registrada y con el argumento `col_wrap` le decimos que represente cuatro gráficos por fila (cuatro columnas).

```python
g = sns.FacetGrid(aux3, col = 'MONTH', margin_titles = True, col_wrap = 4)
bins = np.linspace(aux3['value'].min(), aux3['value'].max(), 25)
g.map(plt.hist, 'value', color = 'steelblue', bins = bins, lw = 0)
g.set_titles(col_template = 'Mes: {col_name}');
```

Este mismo gráfico podemos realizarlo considerando la distribución de `TMIN` y `TMAX`.

```python
g = sns.FacetGrid(aux3, col = 'MONTH', margin_titles = True, col_wrap = 4, hue = 'variable')
bins = np.linspace(aux3['value'].min(), aux3['value'].max(), 25)
g.map(plt.hist, 'value', bins = bins, lw = 0, alpha = 0.6)
g.set_titles(col_template = 'Mes: {col_name}')
g.add_legend(title = 'Temperatura');
```

Al mismo gráfico le añadimos un título y además ponemos la leyenda debajo del gráfico.

```python
g = sns.FacetGrid(aux3, col = 'MONTH', margin_titles = True, col_wrap = 4, hue = 'variable')
bins = np.linspace(aux3['value'].min(), aux3['value'].max(), 25)
g.map(plt.hist, 'value', bins = bins, lw = 0, alpha = 0.6)
plt.subplots_adjust(top = 0.90, bottom = 0.1)
g.fig.suptitle('Histograma mensual')
g.set_titles(col_template = 'Mes: {col_name}')
g.fig.legend(handles = g._legend_data.values(),\
        labels = g._legend_data.keys(),\
        loc = 'lower center', ncol = 2);
```

Con los siguientes comandos podemos conocer la información que maneja el gráfico para representar la leyenda:

  * g._legend_data.values()
  * g._legend_data.keys()

```python
print(g._legend_data.values())
print(g._legend_data.keys())
```

También podemos modificar los ejes X e Y de cada gráfico.

```python
g = sns.FacetGrid(data, col = 'MONTH', margin_titles = True, col_wrap = 4, hue = 'WSF2')
g.map(plt.scatter, 'TMIN', 'TMAX', lw = 3)
plt.subplots_adjust(top = 0.90, bottom = 0.1)
g.fig.suptitle('Scatter plot mensual')
g.set_titles(col_template = 'Mes: {col_name}')
g.add_legend()
g.axes[8].set_xlabel('Temperatura Mínima')
g.axes[9].set_xlabel('Temperatura Mínima')
g.axes[10].set_xlabel('Temperatura Mínima')
g.axes[11].set_xlabel('Temperatura Mínima')

g.axes[0].set_ylabel('Temperatura Máxima')
g.axes[4].set_ylabel('Temperatura Máxima')
g.axes[8].set_ylabel('Temperatura Máxima')
```
