```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6},
                   'D': {0: 1, 1: 1, 2: 1}})
print("\n\ndf")
print("----------------------------------------------------------------------------------------------")
print(df)
print("\n\npd.melt(frame = df)")
print("----------------------------------------------------------------------------------------------")
print(pd.melt(frame = df))
print("\n\npd.melt(frame = df, id_vars = ['A'], value_vars = ['B'], value_name = 'Mis valores')")
print("----------------------------------------------------------------------------------------------")
print(pd.melt(frame = df, id_vars = ['A'], value_vars = ['B'], value_name = 'Mis valores'))
print("\n\npd.melt(frame = df, id_vars = ['A', 'C'], value_vars = ['B'], value_name = 'Mis valores')")
print("----------------------------------------------------------------------------------------------")
print(pd.melt(frame = df, id_vars = ['A', 'C'], value_vars = ['B'], value_name = 'Mis valores'))
print("\n\npd.melt(frame = df, id_vars = ['A', 'D'], value_vars = ['B', 'C'], value_name = 'Mis valores')")
print("----------------------------------------------------------------------------------------------")
print(pd.melt(frame = df, id_vars = ['A', 'D'], value_vars = ['B', 'C'], value_name = 'Mis valores'))
print("----------------------------------------------------------------------------------------------")
print(pd.melt(frame = df, id_vars = ['A', 'D'], value_vars = ['B', 'C'], 
              var_name = 'Mis Variables', value_name = 'Mis Valores'))
```

```python
from numpy.random import randn
dfq = pd.DataFrame(randn(10, 4), columns = list('abcd'))
dfq['pro'] = dfq['a']*dfq['b']*dfq['c']*dfq['d']
dfq['sum'] = dfq['a']+dfq['b']+dfq['c']+dfq['d']
dfq.query('a*b*c*d > a+b+c+d')
```

```python
dfq['sumabs'] = dfq['sum'].apply(abs)
dfq
```

```python
def mifuncion(dt, a, b):
    return dt[a] + dt[b]

dfq.apply(mifuncion, a = 'a', b = 'b', axis = 1)
```

```python
from sklearn.datasets import load_iris

data = load_iris(return_X_y = True)
```

```python
data = pd.concat([pd.DataFrame(data[0], 
                               columns =['a', 'b', 'c', 'd']), 
                  pd.DataFrame(data[1], 
                               columns =['Species'])], 
          axis = 1)
```

```python
data.iloc[0:10]
```

```python
# Ambas opciones son igualemente válidas, la única diferencia es que la primera opción nos 
# dejaría reemplazar variables con nombres con espaciones en blanco, por ejemplo 'mis especies de flores'

#data['Species'].replace({0: 'Iris Setosa', 
#                             1: 'Iris Virginica', 
#                             2: 'Iris Versicolor'}, inplace = True)

data.replace(dict(Species = {0: 'Iris Setosa', 
                             1: 'Iris Virginica', 
                             2: 'Iris Versicolor'}), 
             inplace = True) # Se guarda en el propio dataframe
data.iloc[0:10]
```

```python
data.sort_values('a', ascending = False).iloc[0:10]
```

```python
data.sort_values(['a', 'b'], ascending = False).iloc[0:10]
```

```python
data.rename(columns = {'a' : 'SLength', 
                       'b' : 'SWidth', 
                       'c' : 'PLength', 
                       'd' : 'PWidth'}, inplace = True)
```

```python
data.loc[data['Species'] == 'Iris Setosa', 'Prueba'] = 100 * data['SLength']
data.loc[data['Species'] == 'Iris Versicolor', 'Prueba'] = 10 / data['SLength']
data.loc[data['Species'] == 'Iris Virginica', 'Prueba'] = data['SLength'] * data['SLength']
data.loc[1:10,]
```

```python
import numpy as np
data.loc[:, 'SLength':'PLength'] = np.log(data.loc[:, 'SLength':'PLength'])
data.loc[1:10, ]
```

```python
df.\
    groupby(['asset', 'variable'])['value'].\
    agg(['mean', 'sum', 'count']).\
    rename(columns = {'mean': 'media', 'sum': 'suma', 'count': 'N'},
           inplace  = False)
# Ponemos inplace = False, porque después vamos a volver a cambiar 
# las columnas de nombre en un ejemplo más completo y fallaría.
```

```python
df.\
    groupby(['asset', 'variable'])['value'].\
    agg(['mean', 'std', 'sum', 'count']).\
    rename(columns={'mean': 'media', 'std': 'sd', 'sum': 'suma', 'count': 'N'}).\
    loc[:, ['media', 'sd']].\
    reset_index().\
    assign(CV = lambda x: x.sd/x.media).\
    query('CV < 0.2').\
    head(50)
```
